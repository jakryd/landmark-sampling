```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def v(x, n_terms, sf, x0, sigma):
    val = np.zeros_like(x)
    for i in range(n_terms):
        val += -sf[i] * np.exp(-np.power(x - x0[i], 2) / sigma[i])
    val -= val.min()
    return val
```

```python
x = np.linspace(0, 10, 1000)
val = 50.0 * v(x, n_terms=3, sf=[2.0, 1.2, 1], x0=[2., 5., 8.], sigma=[1., 3.0, 1.]) + 2*(x-5)**2

plt.plot(x, val)
plt.xlim([0,10])
```
